## 10.5.4 支持向量机分类在sklearn中的实现

```python
from sklearn.svm import SVC
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
dataFile = '/Users/lirenyuan/Desktop/maitian/course/data/breast_cancer_m.csv'
data = pd.read_csv(dataFile)
X = data.iloc[: ,0:-1]
y = data['label']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3,random_state=21)
```

```python
svc = SVC(kernel='rbf',probability=True)
svc.fit(X_train,y_train)
```

```python
print(svc.score(X_train,y_train))
print(svc.score(X_test,y_test))
```

```python
svc.predict(X_test)
```

```python
svc.predict_proba(X_test)[:,1]
```

## 10.5.5 网格搜索调参

```python
from sklearn.model_selection import GridSearchCV
```

```python
Cs = np.logspace(-1,3,10,base = 2)
gammas = np.logspace(-5,1,10)
param_grid = {'C': Cs, 'gamma': gammas, 'kernel': ('rbf','linear')}
GS = GridSearchCV(SVC()
                    , param_grid = param_grid
                    , cv = 5
                   )
GS.fit(X_train,y_train)
print(GS.best_params_)
C = GS.best_params_['C']
gamma = GS.best_params_['gamma']
kernel = GS.best_params_['kernel']
C, gamma, kernel
```

```python
C = 0.6803950000871885
gamma = 0.00021544346900318823 
kernel = 'rbf'
```

```python
svc = SVC(kernel=kernel
          , C = C
          , gamma = gamma
          )
svc.fit(X_train,y_train)
print(svc.score(X_train,y_train))
print(svc.score(X_test,y_test))
```

## 10.5.6 p次k折交叉验证在sklearn中的实现

```python
from sklearn.model_selection import RepeatedKFold
```

```python
rkf = RepeatedKFold(n_splits=5, n_repeats=10, random_state=22)
scoreTrainList, scoreTestList = [], []
for train_index, test_index in rkf.split(X):
#     print("TRAIN:", train_index, "TEST:", test_index)
    X_train, X_test = X.iloc[train_index], X.iloc[test_index]
    y_train, y_test = y.iloc[train_index], y.iloc[test_index]
    print(svc.score(X_train,y_train),svc.score(X_test,y_test)) 
    scoreTrainList.append(svc.score(X_train,y_train))
    scoreTestList.append(svc.score(X_test, y_test))
```

```python
print(np.mean(scoreTestList),np.std(scoreTestList))
```
